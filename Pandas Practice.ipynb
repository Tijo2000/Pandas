```python
#Pandas Exercise
import pandas as pd
diamonds = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/diamonds.csv')
print(diamonds)
```

```python
diamonds.head(5)
```

```python
First_6_rows = diamonds[['carat', 'cut', 'x', 'y', 'z']]
First_6_rows.head(7)
```

```python
#Write a Pandas program to create a new 'Quality -color' Series (use bracket notation to define the Series name) of the diamonds DataFrame.
diamonds['Quality-color']=diamonds.cut + ', ' + diamonds.color
diamonds.head(6)
```

```python
#Write a Pandas program to find the number of rows and columns and data type of each column of diamonds Dataframe.
print(diamonds.shape)
print(diamonds.dtypes)
```

```python
#Write a Pandas program to summarize only 'object' columns of the diamonds Dataframe.
print(diamonds.describe(include=['object']))
```

```python
diamonds.rename(columns={'color':'diamond-color','clarity':'diamond-clarity'},inplace = True)
diamonds.head(2)
```

```python
#Write a Pandas program to remove the second column of the diamonds Dataframe.
diamonds.drop('cut',axis = 1, inplace = True)
diamonds
```

```python
diamonds.drop(['depth', 'price'], axis=1, inplace=True)
print(diamonds.head())
```

```python
#Write a Pandas program to remove multiple rows at once (axis=0 refers to rows) from diamonds dataframe.
diamonds.drop([2,3],axis = 0 , inplace = True)
diamonds
```

```python
#Write a Pandas program to sort the 'cut' Series in ascending order (returns a Series) of diamonds Dataframe.
print("\n carat Series in ascending order :")
result = diamonds.carat.sort_values(ascending=True)
print(result)
```

```python
result = diamonds.carat.sort_values(ascending=False)
print(result)
```

```python
result = diamonds.sort_values(by = 'carat', ascending=False)
result
```

```python
#Write a Pandas program to filter the DataFrame rows to only show carat weight at least 0.3.
filt = (diamonds['carat'] >= 0.3)
diamonds =  diamonds[filt]
diamonds
```

```python
result = diamonds[(diamonds.x>5) & (diamonds.y>5) & (diamonds.z>5)]
print(result.head())
```

```python
#Write a Pandas program to find the diamonds that are either Premium or Ideal.
resukt = diamonds[(diamonds.cut=='Premium')|(diamonds.cut=='Ideal')]
resukt
```
